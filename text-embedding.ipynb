```python
import os
import json
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()
client = OpenAI()
```

```python
with open("./data/processed/enriched_chunks.json", "r", encoding="utf-8") as f:
    data_chunks = json.load(f)
    
data_chunks[0]
```

```python
text_chunk = data_chunks[0]['chunk']
context = data_chunks[0]['chunk_metada']['context_expansion']
print(text_chunk)
print(10*'-')
print(context)
```

```python
text = text_chunk.replace('\n', '') + '\n' + context
print(text)
```

```python
def get_embedding(text: str):
    """Generate embedding using openai text-embedding-3-small"""
    
    response = client.embeddings.create(
        input=text,
        model="text-embedding-3-small"
    )
    return response.data[0].embedding
```

```python
embedding = get_embedding(text)
```
